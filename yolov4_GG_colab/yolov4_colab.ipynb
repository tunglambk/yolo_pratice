train.txt: list of files for training

val.txt: list of files for validating

```python
%cd /content/gdrive/My\ Drive/darknet
```

```python
import glob2
import math
import os
import numpy as np
```

```python
files = []

for ext in ['*.png', '*.jpeg', '*.jpg']:
  image_files = glob2.glob(os.path.join("data/", ext))
  files += image_files
```

```python
rand_idx = np.arange(len(files))

np.random.shuffle(rand_idx)

nb_val = math.floor(len(files)*0.2)

idx_val = rand_idx[0:nb_val]
idx_train = rand_idx[nb_val:]
```
